# sorting

```python
df=df.sort_values(by='goals scored', ascending=False)
```

## now make a chart

## long version

```python
import matplotlib.pyplot as plt
plt.subplots(figsize=(10,6))
bars=plt.bar(x='player name',height='goals scored', color='pink', hatch='//',edgecolor='white', width=0.5,data= df.head(10))
# name of country on each bar
for bar,nation in zip(bars,df.head(10).nationality):
    yval = bar.get_height()
    plt.text(x=bar.get_x(),
             y= yval+.1,
             s= nation,
             fontdict={'size':6},
             )
plt.xticks(rotation=45)
plt.title('Top 10 Goals Scorers in 2022')
plt.show()
```

```python
# club with most players
df['club'].value_counts().head(10).plot(kind='pie',
                                        color='orange',
                                        autopct='%.1f%%', # autopct = ->automatic percentage and another % for percentage size
                                        startangle=90,) 
```

```python
# culb with most goals
bars=df.groupby('club')['goals scored'].sum().sort_values(
    ascending=False).head(10).plot(kind='bar')

```

```python
# culb with most goals
bars=df.groupby('club')['goals scored'].sum().sort_values(
    ascending=False).head(10).plot(kind='bar')
# highlight the 1st place
bars.patches[0].set_color('orange')
# goals scored on each bar
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(x= bar.get_x()+ bar.get_width()/2,
             y=1,
             s=f'{int(yval)}',
             ha='center',
             fontdict={'size':12,'color':'white','weight':'bold'},
             rotation=90,
             )
```
